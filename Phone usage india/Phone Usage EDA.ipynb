```python

# Selecting only numerical columns
numerical_cols = ['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)', 
                  'Calls Duration (mins/day)', 'Number of Apps Installed', 
                  'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)', 
                  'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)', 
                  'Monthly Recharge Cost (INR)']

# Creating boxplots for all numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 4, i)  # Adjust grid size based on number of features
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
#numerical variable distribution

plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
#categorical variable distribution

categorical_cols = ['Gender', 'Location', 'Phone Brand', 'OS', 'Primary Use']

plt.figure(figsize=(15, 8))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2, 3, i)
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette="viridis")
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
Age Groups vs. Screen Time & Social Media Usage

df['Age Group'] = pd.cut(df['Age'], bins=[10, 18, 25, 35, 50, 100], labels=['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior'])

plt.figure(figsize=(12,6))
sns.boxplot(x='Age Group', y='Screen Time (hrs/day)', data=df, palette="coolwarm")
plt.title("Screen Time Across Age Groups")
plt.show()
```

```python
# Primary Use vs. Social Media & Streaming

plt.figure(figsize=(12,6))
sns.boxplot(x='Primary Use', y='Social Media Time (hrs/day)', data=df, palette="viridis")
plt.xticks(rotation=45)
plt.title("Social Media Usage by Primary Use")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x=df['Screen Time (hrs/day)'], y=df['Data Usage (GB/month)'], scatter_kws={'alpha':0.5})
plt.title("Screen Time vs. Data Usage")
plt.show()

#A regular scatter plot may show noisy data, so adding a regression line helps detect hidden trends.
#Key questions:
#Does more screen time actually mean more data usage?
#Are there outliers (high screen time, low data usage)?
```

```python
plt.figure(figsize=(8,6))
plt.hexbin(df['Screen Time (hrs/day)'], df['Data Usage (GB/month)'], gridsize=30, cmap='Blues', mincnt=1)
plt.colorbar(label="Density")
plt.xlabel("Screen Time (hrs/day)")
plt.ylabel("Data Usage (GB/month)")
plt.title("Hexbin Plot: Screen Time vs. Data Usage")
plt.show()

#Hexbin plots are useful for large datasets where scatter plots might be too dense.
#Helps identify hotspots where most users lie.
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Social Media Time (hrs/day)'], y=df['E-commerce Spend (INR/month)'], alpha=0.5)
sns.regplot(x=df['Social Media Time (hrs/day)'], y=df['E-commerce Spend (INR/month)'], scatter=False, color='red')
plt.title("Social Media Usage vs. E-Commerce Spending")
plt.show()

#Does more time on social media lead to higher e-commerce spending?
#Are there extreme spenders despite low social media use
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Primary Use', y='Data Usage (GB/month)', data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Data Usage Distribution Across Primary Use Categories")
plt.show()

#Unlike boxplots, violin plots show the full data distribution and density.
#Interesting Questions:
#Which category has widest spread in data usage?
#Which category is the most consistent?
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df[df['Age'] < 25]['E-commerce Spend (INR/month)'], label="Under 25", shade=True)
sns.kdeplot(df[df['Age'] >= 25]['E-commerce Spend (INR/month)'], label="25 & Above", shade=True)
plt.title("Spending Behavior Across Age Groups")
plt.legend()
plt.show()


#KDE plots smooth out spending patterns.
#Helps us see spending peaks and trends by age group.
```

```python
df['Age Group'] = pd.cut(df['Age'], bins=[10, 18, 25, 35, 50, 100], labels=['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior'])

plt.figure(figsize=(10,6))
sns.boxplot(x='Age Group', y='Screen Time (hrs/day)', data=df, palette="viridis")
plt.title("Screen Time Across Age Segments")
plt.show()

#Creating Meaningful Age Segments
```
